## Ejercicio 1

- Cuáles son los vectores de soporte y cuáles de ellos están sobre el margen.    

Los vectores de soporte son aquellos cuyo $ \alpha_i > 0$, de modo que los vectores de soporte son:

1, 2, 4, 5, 6, 7, 8, 9, 10, 11 y 14.

Los vectores de soporte que están sobre el margen son aquellos cuyo $\alpha_i$ está comprendido en $ 0 < \alpha_i < C$ por lo que, dado que $C = 1$, las observaciones cuyos vectores de soporte están sobre el margen son:

2, 5 y 9. 

- Cuáles son los coeficientes del hiperplano ( $ \beta$  y  $ \beta_0 $ ) y el valor de M.

```python
ob0 = np.matrix ([3,2])
ob1 = np.matrix ([1,1])
ob2 = np.matrix ([8,9])
ob3 = np.matrix ([-1,0])
ob4 = np.matrix ([1,-2])
ob5 = np.matrix ([0,-1])
ob6 = np.matrix ([1,-3])
ob7 = np.matrix ([-1,-3])
ob8 = np.matrix ([-5,-4])
ob9 = np.matrix ([-2,-5])
ob10 = np.matrix ([1,-7])
ob11 = np.matrix ([-9,-8])

#Calculamos los vectores de Beta
beta = -1*0.125*ob0+ob1-ob3-ob4+ob5-ob6+ob7+0.31*ob8-ob9+0.82*ob10

print beta
```

Una vez calculados los valores de $ \beta $ podemos calcular $ \beta_0$ fácilmente utilizando la fórmula $ (1)$ de la transparencia 14 de los apuntes, para cualquier vector de soporte que está en el borde, lo hacemos para las observaciones 2 y 9 y hacemos que $ \beta_0 $ sea la media de las dos, y obtenemos:

```python
beta0_ob2 = 0.92
beta0_ob5 = 0.86
beta0_ob9 = 0.92

#Calculamos Beta0 aplicando la fórmula antes mencionada
beta0 = (beta0_ob2 + beta0_ob5 + beta0_ob9)/3

print beta0
```

Por último hacemos la inversa sobre la norma de beta para hallar $M$: $$M = \frac{1}{||\beta||}$$

```python
M = 1/np.linalg.norm(beta) #Podemos aprovechar la librería numpy para calcular la norma en lugar de hacerlo a mano
print M
```

- Los valores de $\epsilon_i$ y las observaciones incorrectamente clasificadas

De las fórmulas $(5)$ y $(2)$ deducimos fácilmente que $\epsilon_i \neq 0$ si y sólo si $ \alpha = C$, por lo que solo tendremos que calcular dicho valor para las observaciones: 1, 4, 6, 7, 8, 10, 11, 14.

Para el resto $\epsilon_i = 0$.

Para calcular $\epsilon_i$ utilizamos la fórmula $(1)$ y obtenemos:

- $\epsilon_1 = 0.53$
- $\epsilon_4 = 1.51$
- $\epsilon_6 = 1.98$
- $\epsilon_7 = 0.43$
- $\epsilon_8 = 1.43$
- $\epsilon_{10} = 0.51$
- $\epsilon_{11} = 2.06$
- $\epsilon_{14} = 0.06$

Por último, para ver que observaciones estarían mal clasificadas utilizamos $$G(x)=sign[x^T\beta + \beta_0]$$

```python
def clasificacion(v):
    resultado = v * np.transpose(beta) + beta0
    return resultado 

print "Vectores clasificados como 1"
print "clasificación o0 = sign", clasificacion(o0)
print "clasificación o1 = sign", clasificacion(o1)
print "clasificación o2 = sign", clasificacion(o2)
print "clasificación o3 = sign", clasificacion(o3)
print "clasificación o4 = sign", clasificacion(o4)
print "clasificación o5 = sign", clasificacion(o5)
print "clasificación o6 = sign", clasificacion(o6)
print "clasificación o7 = sign", clasificacion(o7)
```

Por tanto, de los vectores clasificados como 1, estarían mal clasificados el 4 y el 6.

```python
print "Vectores clasificados como -1"
print "clasificación o8 = sign", clasificacion(o8)
print "clasificación o9 = sign", clasificacion(o9)
print "clasificación o10 = sign", clasificacion(o10)
print "clasificación o11 = sign", clasificacion(o11)
```

- random_state : int seed, RandomState instance, or None (default)

### Apartado b

```python
datasetName = 'bloodTransfusion.data'
datasetDelimiter = ','

# Cargamos los datos
dataset = np.loadtxt(datasetName, delimiter=datasetDelimiter)
xRaw = dataset[:,0:dataset.shape[1]-1] 
y = dataset[:,dataset.shape[1]-1]
#Si hay valores perdidos o 'NaN' los imputamos
imp = Imputer(missing_values='NaN', strategy='mean', axis=0)
xPrep = imp.fit_transform(xRaw)
#Estandarizamos los datos:
scaler = preprocessing.StandardScaler().fit(xRaw)
x=scaler.transform(xRaw)

xTrain, xTest, yTrain, yTest = cross_validation.train_test_split(x, y, test_size=0.2, random_state=0)
print 'xTrain', len(xTrain), "\nxTest", len(xTest), "\nyTrain", len(yTrain), "\nyTest", len(yTest)
```

### Apartado c

```python
#Error rbf
for i in range(1,6):
    hyperParams = {'kernel': ['rbf'], 'C': [0.001,0.01,0.1,1,10], 'gamma': [i]}
    modelCV = GridSearchCV(svm.SVC(),hyperParams, cv=5)
    modelCV.fit(xTrain, yTrain)
    print "Best hyperparameters", modelCV.best_params_
    c, errList, devList = [], [], []
    for hyperP, mean_score, scores in modelCV.grid_scores_:
        c.append(hyperP['C'])
        errList.append(1-mean_score)
        devList.append(scores.std())
        if hyperP['C'] == modelCV.best_params_['C'] and hyperP['gamma'] == modelCV.best_params_['gamma']:
            print "Error: " + str(1-mean_score)
    plt.errorbar(c, errList)
    plt.show()
```

```python
#Error polinomico
for i in range(2,5):
    hyperParams = {'kernel': ['poly'], 'C': [0.0001, 0.001,0.01,0.1,1],'degree':[i]}
    #Create an instance of Neighbors Classifier and fit the data for the grid parameters
    modelCV = GridSearchCV(svm.SVC(),hyperParams, cv=5)
    modelCV.fit(xTrain, yTrain)
    print "Best hyperparameters", modelCV.best_params_
    c, errList, devList = [], [], []
    for hyperP, mean_score, scores in modelCV.grid_scores_:
        c.append(hyperP['C'])
        errList.append(1-mean_score)
        devList.append(scores.std())
        if hyperP['C'] == modelCV.best_params_['C'] and hyperP['degree'] == modelCV.best_params_['degree']:
            print "Error: " + str(1-mean_score)
    plt.errorbar(c, errList)
    plt.show()
```

```python
#Error lineal
hyperParams = {'kernel': ['linear'], 'C': [0.001,0.01,0.1,1,10, 100]}
modelCV = GridSearchCV(svm.SVC(),hyperParams, cv=5)
modelCV.fit(xTrain, yTrain)
print "Best hyperparameters", modelCV.best_params_
c, errList, devList = [], [], []
for hyperP, mean_score, scores in modelCV.grid_scores_:
    c.append(hyperP['C'])
    errList.append(1-mean_score)
    devList.append(scores.std())
    if hyperP['C'] == modelCV.best_params_['C']:
          print "Error: " + str(1-mean_score)
plt.errorbar(c, errList)
plt.show()
```

```python
numC,errList2 = [],[]
for j in range (-4,3):
    c=10**j
    clf = svm.SVC(kernel= 'linear',C =c)
    clf.fit(xTrain, yTrain)
    clf.predict(xTest)
    error=100-clf.score(xTest,yTest)*100
    errList2.append(error)
    print "Mínimo error de clasificación:",min(errList2)
    numC.append(c)
plt.errorbar(numC, errList2)
plt.show()
```

```python
for i in range (1,6):
    numC,errList2 = [],[]
    for j in range (-3,3):
        c=10**j
        clf = svm.SVC(kernel= 'rbf',C =c,gamma=i)
        clf.fit(xTrain, yTrain)
        clf.predict(xTest)
        error=100-clf.score(xTest,yTest)*100
        errList2.append(error)
        print "Mínimo error de clasificación:",min(errList2)
        numC.append(c)
    plt.errorbar(numC, errList2)
    plt.show()
```

```python
for i in range (2,5):
    numC,errList2 = [],[]
    for j in range (-4,1):
        c=10**j
        clf = svm.SVC(kernel= 'poly',C =c,degree=i)
        clf.fit(xTrain, yTrain)
        clf.predict(xTest)
        error=100-clf.score(xTest,yTest)*100
        errList2.append(error)
        print "Mínimo error de clasificación:",min(errList2)
        numC.append(c)
    plt.errorbar(numC, errList2)
    plt.show()
```
