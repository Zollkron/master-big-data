```python
import numpy as np
#Observaciones
matrix = np.array([[2,6,1,0],
                   [4,3,1,1],
                   [4,4,1,0.33],
                   [4,6,1,0],
                   [6,3,1,1],
                   [7,7,1,0.17],
                   [8,4,1,1],
                   [9,8,1,1],
                   [2,1,-1,1],
                   [6,2,-1,0.5],
                   [7,4,-1,1],
                   [8,8,-1,1],
                   [9,1,-1,0],
                   [10,3,-1,0],
                   [10,6,-1,1],
                   [12,4,-1,0]])
#Array vector de soporte y si es un vector de soporte en el margen
C=1
suppVect = []
suppMargVect = []
outsideMargin=[]
for i in range(0,len(matrix)):
    if (matrix[i, 3]==0):
        print("El vector ",i, " está fuera del margen")
        outsideMargin.append(i)
    elif (matrix[i, 3]==1):
        print("El vector ", i, " es un VECTOR DE SOPORTE")
        suppVect.append(i)
    elif ((matrix[i,3]>0) and (matrix[i,3]<C)):
        print("El vector ", i, " es un VECTOR DE SOPORTE  y ESTÁ DENTRO DEL MARGEN")
        suppMargVect.append(i)
        suppVect.append(i)
beta = np.array([0,0])
for i in suppVect:
    x = np.array([matrix[i][0],matrix[i][1]])
    y = matrix[i][2]
    alpha = matrix[i][3]
    beta = beta + alpha * y *x
    
print("Beta = ",beta)
print()
beta0 = 0
for i in suppMargVect:
    x = np.array([matrix[i][0],matrix[i][1]])
    y = matrix[i][2]
    alpha = matrix[i][3]
    beta0 = beta0 + (1 - (np.dot(x.transpose(),beta)*y)) / y
#Ahora hacemos la media para obtener beta0
beta0 = beta0/len(suppMargVect)
    
print()

print("Beta 0 = ",beta0)
    
#M = 1/||beta||
M = 1/(np.linalg.norm(beta))
print("M = ",M)
epsilon= []


for i in suppVect:
    x = np.array([matrix[i][0],matrix[i][1]])
    y = matrix[i][2]
    actualEpsilon = 1-y*(np.dot(x.transpose(), beta)+beta0)
    if (actualEpsilon<0):
        epsilon.append(round(-actualEpsilon,2))
    else : epsilon.append(round(actualEpsilon,2))

print("Hemos obtenido las siguientes epsilon:")
[print(round(x,2))  for x in epsilon]
print()


#Definimos la función G para obtener el valor de la clasificación. Si G==0 
def g(x, beta, beta0):
    return np.dot(x.transpose(),beta)+beta0

for i in range (len(matrix)):
    x = np.array([matrix[i][0],matrix[i][1]])
    actualG = g(x,beta,beta0)
   
    if ((actualG > 0) and (matrix[i][2] !=1)):
        print ("La observacion " ,i, " esta mal clasificada con G(x)=", actualG," y está clasificada como: ",matrix[i][2])
    elif((actualG<0) and (matrix[i][2]!=-1)):
        print ("La observacion " ,i, " esta mal clasificada: G(x)=", actualG," y está clasificada como:",matrix[i][2])
```
